```python
ts_client = typesense.Client({
    'api_key': 'wxsi1sbihy6f5hh6',
    'nodes': [{
        'host': 'search.absclust.com',
        'port': '443',
        'protocol': 'https'
    }],
    'connection_timeout_seconds': 2
})


app_config = {"TYPESENSE_PERPAGE": 100,
              "TYPESENSE_MAXPAGES": 20}
```

```python
def get_docs(q, query_by="abstract, title", filter_by=None, sort_by=None, **kwargs):
    col = ts_client.collections["articles"]
    res = []
    params = {"q": q, "query_by": query_by, "filter_by": filter_by, "sort_by": sort_by}
    params = {k: v for k, v in params.items() if v is not None}
    hits = col.documents.search(
        dict(
            **params,
            per_page=app_config["TYPESENSE_PERPAGE"],
            highlight_fields="",
            page=1,
        )
    )
    res.extend(hits["hits"])
    max_pages = math.ceil(hits["found"] / app_config["TYPESENSE_PERPAGE"])
    max_page = min([max_pages, app_config["TYPESENSE_MAXPAGES"]])
    if max_page > 1:
        for page in range(2, max_page + 1):
            hits = col.documents.search(
                dict(
                    **params,
                    per_page=app_config["TYPESENSE_PERPAGE"],
                    highlight_fields="",
                    page=page,
                )
            )
            res.extend(hits["hits"])
    return [a["document"] for a in res]
```
